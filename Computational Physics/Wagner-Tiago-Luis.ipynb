```python
import math as mt
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

# Estimativa de par√¢metros

```python
L = np.array([14.8, 11.8, 8.8, 5.8])
T = np.array([0.8, 0.72, 0.61, 0.51])
SigmaT = np.array([.01,.01,.01,.01])
SigmaT2 = 2*T*SigmaT
```

```python
def f(x,a,b):
    return a*x+b

[[a,b], matriz] = curve_fit(f, L, T, np.array([1,1]), SigmaT)

print('a = ', 4*np.pi/9.81)

plt.plot(L, f(L, a, b), 'b')
plt.plot(L, T, 'b.')
plt.legend(['Estimative', 'Measured Values'])
plt.xlabel('L (cm)')
plt.ylabel('T (s)')
plt.show()
```
